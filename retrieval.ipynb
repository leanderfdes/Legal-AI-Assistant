```python
import chromadb
from sentence_transformers import SentenceTransformer

# Load ChromaDB
chroma_path = "./new_vector_db"  # Path where we saved the vector database
chroma_client = chromadb.PersistentClient(path=chroma_path)
collection = chroma_client.get_collection(name="ipc_sections")

# Load the same embedding model used for storing
embedding_model = SentenceTransformer("./models/all-MiniLM-L6-v2")

def retrieve_ipc_sections(query, top_k=3):
    """Find the most relevant IPC sections for a given query."""
    query_embedding = embedding_model.encode(query).tolist()

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )

    if results["ids"]:
        return [
            f"Section {meta['section']}: {meta['description']}"
            for meta in results["metadatas"][0]
        ]
    return ["No relevant IPC section found."]

# Example usage
query = "i harassed my wife"
retrieved_sections = retrieve_ipc_sections(query)
print("\n".join(retrieved_sections))
```
